```python
timesData = pd.read_csv("/content/timesData.csv")
```

```python
timesData.head()
```

```python
df2012 = timesData[timesData.year == 2012].iloc[:4,:]
```

```python
trace1 = go.Bar( 
                x = df2012.university_name, 
                y = df2012.citations,  
                name = "citations",    
                marker = dict(color = 'rgba(255, 174, 255, 0.5)', 
                             line=dict(color='rgb(0,0,0)',width=1.5)), 
                text = df2012.country) 

trace2 = go.Bar(
                x = df2012.university_name,  
                y = df2012.teaching,    
                name = "teaching",     
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',  
                              line=dict(color='rgb(0,0,0)',width=1.5)), 
                text = df2012.country)   

data = [trace1, trace2] 

layout = go.Layout(barmode = "group") 
fig = go.Figure(data = data, layout = layout) 

iplot(fig)
```

```python
x = df2012.university_name

trace1 = {
  'x': x,
  'y': df2012.citations,
  'name': 'citation', 
  'type': 'bar' 
};
trace2 = {
  'x': x,  
  'y': df2012.teaching,  
  'name': 'teaching',   
  'type': 'bar'  
};

data = [trace1, trace2]; 

layout = {
  'xaxis': {'title': 'Top 4 universities'},
  'barmode': 'relative',  
  'title': 'citations and teaching of top 4 universities in 2012'
};

fig = go.Figure(data = data, layout = layout) 

iplot(fig) 
```
