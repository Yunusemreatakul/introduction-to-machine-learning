```python
timesData = pd.read_csv("/content/timesData.csv")
```

```python
timesData.head()
```

```python
df2014 = timesData[timesData.year == 2014].iloc[:100,:] 
df2012 = timesData[timesData.year == 2012].iloc[:100,:]
df2013 = timesData[timesData.year == 2013].iloc[:100,:]
```

```python
trace1 =go.Scatter( 
                    x = df2014.world_rank, 
                    y = df2014.citations,  
                    mode = "markers",     
                    name = "2014",         
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'), 
                    text= df2014.university_name) 


trace2 =go.Scatter(
                    x = df2012.world_rank,
                    y = df2012.citations,  
                    mode = "markers",      
                    name = "2012",         
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'), 
                    text= df2012.university_name)  

trace3 =go.Scatter(
                    x = df2013.world_rank,
                    y = df2013.citations,  
                    mode = "markers",      
                    name = "2013",         
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),  
                    text= df2013.university_name)  

data = [trace1, trace2, trace3]  
layout = dict(title = 'Citation vs world rank of top 100 universities with 2014, 2012 and 2013 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Citation',ticklen= 5,zeroline= False)  
             )
fig = dict(data = data, layout = layout)  

iplot(fig)
```

```python
trace1 =go.Scatter( 
                    x = df2014.world_rank, 
                    y = df2014.total_score,  
                    mode = "markers",     
                    name = "2014",         
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= df2014.university_name)  


trace2 =go.Scatter(
                    x = df2012.world_rank, 
                    y = df2012.total_score,  
                    mode = "markers",      
                    name = "2012",         
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),  
                    text= df2012.university_name)  


trace3 =go.Scatter(
                    x = df2013.world_rank, 
                    y = df2013.total_score,  
                    mode = "markers",      
                    name = "2013",        
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),  
                    text= df2013.university_name)  

data = [trace1, trace2, trace3]  

layout = dict(title = 'Total score vs world rank of top 100 universities with 2014, 2012 and 2013 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Total Score',ticklen= 5,zeroline= False)  
             )
fig = dict(data = data, layout = layout)  

iplot(fig)
```
